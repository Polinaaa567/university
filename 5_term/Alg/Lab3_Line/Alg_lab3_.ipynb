```python
class Node:

    def __init__(self, initdata):
        self.data = initdata
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self, newdata):
        self.data = newdata

    def setNext(self, newnext):
        self.next = newnext


class UnorderedList:

    def __init__(self):
        self.head = None

    # задание №2. реализация метода __str__
    def __str__(self):
        current = self.head
        out = '['
        while current != None:
            out += str(current.getData()) + ', '
            current = current.getNext()
        return out[:-2] + ']' if len(out) > 2 else '[]'

    def isEmpty(self):
        return self.head == None

    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count += 1
            current = current.getNext()

        return count

    def search(self, item):
        current = self.head
        found = False
        while current != None and not found:
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()

        return found

    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.getData() == item:
                found = True
            else:
                previous = current
                current = current.getNext()

        if previous == None:
            self.head = current.getNext()
        else:
            previous.setNext(current.getNext())

    def append(self, item):
        current = self.head
        if current:
            while current.getNext() != None:
                current = current.getNext()
            current.setNext(Node(item))
        else:
            self.head = Node(item)

    def index(self, item):
        current = self.head
        index = 0
        while current:
            if current.getData() == item:
                return index
            else:
                current = current.getNext()
                index += 1
        return None

    def insert(self, pos, item):
        if pos >= self.size() or pos < 0:
            return 'Индекс вне диапазона'
        else:
            if pos == 0:
                self.add(item)
            else:
                current = self.head
                previous = None
                index = 0
                while index < pos:
                    previous = current
                    current = current.getNext()
                    index += 1
                newNode = Node(item)
                newNode.setNext(current)
                previous.setNext(newNode)

    def pop(self, pos=None):
        if pos is None:
            pos = self.size() - 1
        if pos >= self.size() or pos < 0:
            return 'Индекс вне диапазона'
        else:
            current = self.head
            if pos == 0:
                self.head = current.getNext()
                return current.getData()
            else:
                previous = None
                index = 0
                while index < pos:
                    previous = current
                    current = current.getNext()
                    index += 1
                result = current.getData()
                previous.setNext(current.getNext())
                return result

    def slice(self, start, stop):
        if start >= self.size() or start < 0 or stop > self.size() or stop < 0:
            return 'Индекс вне диапазона'
        elif start > stop:
            return 'Недопустимый срез'
        else:
            current = self.head
            index = 0
            slicelist = UnorderedList()
            while index < start:
                current = current.getNext()
                index += 1
            while index < stop:
                slicelist.append(current.getData())
                current = current.getNext()
                index += 1
            return slicelist


mylist = UnorderedList()

# задание №1. проверка методов append(item), index(item), insert(pos,item), pop(), pop(pos)

# append(item)
mylist.append(1)
mylist.append(5)
mylist.append(10)
mylist.append(20)
mylist.append(50)
mylist.append(100)
print(mylist)

print()

# index(item)
print(mylist.index(5))
print(mylist.index(50))
print(mylist.index(100000))

print()

# insert(pos,item)
mylist.insert(1, 88)
print(mylist)
mylist.insert(5, 9999)
print(mylist)
# если писать также, как и два прошлых insert(), список не поменяется, но и не выведется "Индекс вне диапазона"
print(mylist.insert(100, 13))

print()

# pop()
popped_item = mylist.pop()
print(mylist)

print()

# pop(pos)
popped_item = mylist.pop(pos=4)
print(mylist)

print()

# задание №3. проверка среза
slice_list = mylist.slice(1, 4)
print(slice_list)
slice_list = mylist.slice(1, 7)
print(slice_list)
slice_list = mylist.slice(-1, 3)
print(slice_list)
```

Напишите функцию move_to_front для чтения строк с клавиатуры и сохранения их в
связном списке без повторов. Пользователь последовательно вводит с клавиатуры n строк.
При вводе строки, которая раньше не встречалась, вставьте ее в начало списка. При вводе
уже внесенной строки удалите ее из списка и снова вставьте в начало.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self, data):
        if self.head is None:
            return

        if self.head.data == data:
            self.head = self.head.next
            return

        current = self.head
        while current.next is not None:
            if current.next.data == data:
                current.next = current.next.next
                return
            current = current.next

    def move_to_front(self, data):
        self.delete(data)
        self.insert_at_beginning(data)

    def print_list(self):
        current = self.head
        while current is not None:
            print(current.data)
            current = current.next


def move_to_front():
    n = int(input("Введите количество строк: "))
    linked_list = LinkedList()

    for _ in range(n):
        string = input("Введите строку: ")
        linked_list.move_to_front(string)

    linked_list.print_list()


move_to_front()
```

Напишите функцию, которая принимает в качестве аргумента первый узел односвязного
списка, обращает порядок элементов списка (с разрушением исходного) и возвращает в
качестве результата первый узел полученного списка. Разработайте два варианта решения
этой задачи: 1) итеративный и 2) рекурсивный.

```python
class ListNode:
    def __init__(self, data=None, next_node=None):
        self.data = data
        self.next = next_node


def reverse_list_iterative(head):
    prev = None
    curr = head

    while curr:
        new_node = ListNode(curr.data)
        new_node.next = prev
        prev = new_node

        curr = curr.next

    return prev


def reverse_list_recursive(head):
    if not head or not head.next:
        return ListNode(head.data)

    reversed_list = reverse_list_recursive(head.next)
    new_head = ListNode(head.data)
    last_node = reversed_list
    while last_node.next:
        last_node = last_node.next
    last_node.next = new_head

    return reversed_list


def print_list(head):
    curr = head
    while curr:
        print(curr.data, end=" -> ")
        curr = curr.next
    print("None")


node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node1.next = node2
node2.next = node3
node3.next = node4

print("Исходный список:")
print_list(node1)

reversed_head_iterative = reverse_list_iterative(node1)
print("Реверсированный с помощью итеративной функции список:")
print_list(reversed_head_iterative)

reversed_head_recursive = reverse_list_recursive(node1)
print("Реверсированный с помощью рекурсивной функции список:")
print_list(reversed_head_recursive)
```

Реализуйте класс DoubleList для построения двухсвязных списков, в которых каждый узел
содержит ссылку на узел, предшествующий ему в списке, и ссылку на следующий узел.
Класс должен иметь следующие методы: проверка списка на пустоту, поиск элемента,
размер списка, вставка в начало, вставка в конец, удаление из начала, удаление из конца,
вставка перед указанным узлом, вставка после указанного узла, удаление указанного узла.
Проведите тестирование методов реализованного класса.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoubleList:
    def __init__(self):
        self.head = None
        self.tail = None

    def is_empty(self):
        return self.head is None

    def size(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    def insert_at_start(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def delete_from_start(self):
        if self.is_empty():
            print("Список пуст. Удаление из начала невозможно.")
            return
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None

    def delete_from_end(self):
        if self.is_empty():
            print("Список пуст. Удаление из конца невозможно.")
            return
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None

    def insert_before(self, node_value, data):
        new_node = Node(data)
        if self.is_empty():
            print("Список пуст. Невозможно вставить до.")
            return
        current = self.head
        while current:
            if current.data == node_value:
                if current == self.head:
                    self.insert_at_start(data)
                else:
                    new_node.prev = current.prev
                    new_node.next = current
                    current.prev.next = new_node
                    current.prev = new_node
                return
            current = current.next
        print(f"Узел {node_value} не найден.")

    def insert_after(self, node_value, data):
        new_node = Node(data)
        if self.is_empty():
            print("Список пуст. Невозможно вставить после.")
            return
        current = self.head
        while current:
            if current.data == node_value:
                if current == self.tail:
                    self.insert_at_end(data)
                else:
                    new_node.prev = current
                    new_node.next = current.next
                    current.next.prev = new_node
                    current.next = new_node
                return
            current = current.next
        print(f"Узел {node_value} не найден.")

    def delete_node(self, node_value):
        if self.is_empty():
            print("Список пуст. Удаление узла невозможно.")
            return
        current = self.head
        while current:
            if current.data == node_value:
                if current == self.head:
                    self.delete_from_start()
                elif current == self.tail:
                    self.delete_from_end()
                else:
                    current.prev.next = current.next
                    current.next.prev = current.prev
                return
            current = current.next
        print(f"Узел {node_value} не найден.")

    def display(self):
        if self.is_empty():
            print("Список пуст.")
            return
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()


# Использование методов класса DoubleList
dll = DoubleList()

while True:
    print("\n1. Вставка в начало")
    print("2. Вставка в конец")
    print("3. Удаление из начала")
    print("4. Удаление из конца")
    print("5. Вставка до")
    print("6. Вставка после")
    print("7. Удаление узла")
    print("8. Отображение")
    print("9. Размер")
    print("10. Выход")

    option = int(input("Выберите опцию: "))

    if option == 1:
        dll.insert_at_start(int(input("Введите данные: ")))
    elif option == 2:
        dll.insert_at_end(int(input("Введите данные: ")))
    elif option == 3:
        dll.delete_from_start()
    elif option == 4:
        dll.delete_from_end()
    elif option == 5:
        node_value = int(input("Введите существующее значение узла: "))
        data = int(input("Введите новые данные: "))
        dll.insert_before(node_value, data)
    elif option == 6:
        node_value = int(input("Введите существующее значение узла: "))
        data = int(input("Введите новые данные: "))
        dll.insert_after(node_value, data)
    elif option == 7:
        dll.delete_node(int(input("Введите значение узла: ")))
    elif option == 8:
        dll.display()
    elif option == 9:
        print("Размер списка:", dll.size())
    elif option == 10:
        break
    else:
        print("Неверная опция. Пожалуйста, выберите заново.")
```

В HTML у каждого тега существуют открывающая и закрывающая формы, которые должны
быть сбалансированы, чтобы адекватно описывать веб-документ. Напишите программу, которая, используя структуру данных стек, проверит HTML
документ на соответствие открывающих и закрывающих тегов.

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None

    def is_empty(self):
        return len(self.stack) == 0

    def top(self):
        if not self.is_empty():
            return self.stack[-1]
        return None


def is_html_tags_balanced(html_code):
    stack = Stack()
    balanced = True
    index = 0

    while index < len(html_code) and balanced:
        char = html_code[index]

        if char == '<':
            stack.push(char)
        elif char == '>':
            if stack.is_empty():
                balanced = False
            else:
                stack.pop()
        index += 1

    return balanced and stack.is_empty()


html_code = '''<html>
 <head>
 <title>
 Example
 </title>
 </head>
 <body>
 <h1>Hello, world</h1>
 </body>
</html>'''

if is_html_tags_balanced(html_code):
    print("HTML теги сбалансированы.")
else:
    print("HTML теги НЕ сбалансированы.")
```

Задание №7

Реализуйте стековый калькулятор. Напишите программу, которая читает выражение в
обратной польской нотации и выводит получившееся значение или пишет, что выражение
составлено не корректно (если оно некорректно).


```python
def evaluate_expression(expression):
    stack = []

    tokens = expression.split()

    for token in tokens:
        if token.isdigit():
            stack.append(int(token))
        else:
            if len(stack) < 2:
                return "Некорректное выражение"
            operand2 = stack.pop()
            operand1 = stack.pop()

            if token == '+':
                stack.append(operand1 + operand2)
            elif token == '-':
                stack.append(operand1 - operand2)
            elif token == '*':
                stack.append(operand1 * operand2)
            elif token == '/':
                if operand2 == 0:
                    return "Деление на ноль"
                stack.append(operand1 / operand2)
            else:
                return "Некорректный оператор"

    if len(stack) == 1:
        return stack[0]
    else:
        return "Некорректное выражение"


expression = input("Введите выражение (пример формата: '5 1 + 2 *'): ")
result = evaluate_expression(expression)
print(f"Результат: {result}")
```

```python
from queue import Queue
import time


class Store:
    def __init__(self):
        self.queue = Queue()

    def enter_queue(self, customer):
        self.queue.put(customer)
        print(f"{customer} вошел в очередь.")

    def process_queue(self):
        while not self.queue.empty():
            customer = self.queue.get()
            print(f"Обслуживается покупатель: {customer}")
            time.sleep(2)  # имитация времени обслуживания покупателя
            print(f"{customer} завершил покупку и покинул магазин.")

    def get_queue_size(self):
        return self.queue.qsize()


bakery_store = Store()

bakery_store.enter_queue("Покупатель 1")
bakery_store.enter_queue("Покупатель 2")
bakery_store.enter_queue("Покупатель 3")
bakery_store.enter_queue("Покупатель 4")
bakery_store.enter_queue("Покупатель 5")

bakery_store.process_queue()

print(f"Размер оставшейся очереди: {bakery_store.get_queue_size()}")
```
