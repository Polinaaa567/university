№2 (4 балла)
Создайте рисунок с основными элементами изображения, похожий на приведенный ниже. Элементы изображения обозначены кругами и подписаны синим шрифтом. Допускается изменение линий, цвета, толщины линий и т.п., главное, чтобы все обозначенные элементы
присутствовали.
image.png

```python
# 2

plt.figure(figsize=(6, 6))
x1 = np.linspace(0, 4)
y1 = lambda x1: np.cos(x1) + 1

plt.plot(x1, y1(x1), 'r', label='Красная линия - cos(x) + 1')

y2 = lambda x1: np.abs(x1)
plt.plot(x1, y2(x1), 'b', label='Синяя линия - abs(x)')

x1 = np.linspace(0, 1.5)
y3 = lambda x1: x1 ** 3
plt.scatter(x1, y3(x1), color='c')

plt.minorticks_on()
plt.subplot().xaxis.set_minor_formatter(FormatStrFormatter('%.1f'))
plt.tick_params(which="minor",  labelsize=8)

plt.grid( linestyle="--")

plt.yticks([i for i in np.arange(0, 5)])
plt.xticks([i for i in np.arange(0, 5)])

plt.title('Элементы изображения')
plt.xlabel('Подпись оси ОХ')
plt.ylabel('Подпись оси ОY')
plt.legend()
```
