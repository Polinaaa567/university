4. Рассчитайте параметры описательной статистики для признаков набора. С помощью
параметров описательной статистики опишите один из количественных признаков и
один из категориальных признаков. Есть ли в количественном признаке выбросы?
Какое значение является выбросом и почему? По какому категориальному признаку
можно группировать данные? Какие группировки представляют интерес и почему?


```python
df_d = df['duration']

print('Описание для продолжительности времени')
print('Среднее:', np.mean(df_d))
print('Гармоническое среднее:', statistics.harmonic_mean(df_d))
print('Среднее геометрическое:',  statistics.geometric_mean(df_d))
print('Расчет медианы:', statistics.median(df_d))
print('Вычисление моды:', statistics.multimode(df_d))
print('Оценка дисперсии:', statistics.variance(df_d))
print('Среднеквадратичное отклонение:', statistics.stdev(df_d))
print('Смещение:', df_d.skew())
print('Диапазон:', np.ptp(df_d))
print('Сводка описательной статистики:\n ', df_d.describe())
```

6. Проведите корреляционный анализ. Есть ли в вашем наборе данных коррелирующие
признаки?

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 5))

sns.heatmap(corrmat, vmax=.8, square=True, ax=ax)
plt.show()
```

```python
top10_attr = corrmat.nlargest(10, 'duration').index
top10_mat = corrmat.loc[top10_attr, top10_attr]
fig, ax = plt.subplots(figsize=(12, 5))
sns.set(font_scale=1.25)
sns.heatmap(top10_mat, annot=True, annot_kws={'size': 12},
            square=True)
plt.show()
```
